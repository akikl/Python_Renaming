```python
import os
```

```python
folder=r'C:/Users/Dell/Desktop/Python_classes/Pandas/keystone/test'
file_extensions=['.jpg','.png','.pdf','.jpeg']
images=[]
```

```python
for i in os.listdir(folder):
    if '.'+i.rsplit('.',1)[-1].lower() in file_extensions and i not in images:
        images.append(i)
#print(images)
count_images=len(images)
```

```python
document_category=input('enter a document category')
```

```python
document_tye=input('enter a document type')
```

```python
for i in range(1,len(images)+1):
    if i in range(1,10):
        new_image_name=document_category+'-'+document_tye+'-'+'H000'+str(i)
        print(new_image_name)
        os.rename(images[i-1],new_image_name+'.'+(images[i-1]).rsplit('.',1)[-1])
    elif i>99:
        new_image_name=document_category+'-'+document_tye+'-'+'H0'+str(i)
        print(new_image_name)
        os.rename(images[i-1],new_image_name+'.'+images[i-1].rsplit('.',1)[-1])
        
    else:
        new_image_name=document_category+'-'+document_tye+'-'+'H00'+str(i)
        print(new_image_name)
        os.rename(images[i-1],new_image_name+'.'+images[i-1].rsplit('.',1)[-1])
print(f"There are {count_images} images renamed")
```
